```python
from neighbors import NNeighborRegressor
from datasets import make_regression, make_polynomial
from splitters import splitXy
from plots import plot2D, plot_bf_2d
import numpy as np
import time
```

```python
X, y = make_regression(n_samples=1000, n_features=1, noise=0.0, random_state=54)
```

```python
title = "Make Regression Dataset"
xlabel = "Feature 1"
ylabel = "Target 1"
plot2D("2D " + title, xlabel, ylabel, X, y)
```

```python
X_train, X_test, y_train, y_test = splitXy(X, y, test_size=0.2, random_state = 111)
```

```python
start = time.time()
nn = NNeighborRegressor()
nn.fit(X_train, y_train)
score = nn.score(X_test, y_test)
end = time.time()
print("Nearst Neighbor Regressor R2 Score is ", np.round(100 * score, 2), "%")
print('Time taken ', np.round(end-start, 2),' seconds')
```

```python
start = time.time()
title = "Best Fit"
plot_bf_2d(nn, X, y, title, xlabel, ylabel, degree=1)
end = time.time()
print('Time taken ', np.round(end-start, 2),' seconds')
```
