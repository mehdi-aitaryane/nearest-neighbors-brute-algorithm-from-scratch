```python
from neighbors import NNeighborClassifier
from datasets import make_classification, make_blobs, make_circles
from splitters import splitXy
from plots import scatter2D, plot_db_2d
import numpy as np
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=4, noise=0.0, balanced=True, random_state=54)
```

```python
title = "Make Classification Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
scatter2D(title, xlabel, ylabel, X, y)
```

```python
X_train, X_test, y_train, y_test = splitXy(X, y, test_size=0.2, random_state = 111)
```

```python
nn = NNeighborClassifier()
nn.fit(X_train, y_train)
score = nn.score(X_test, y_test)
print("Nearst Neighbor Classifier Accuracy is ", np.round(100 * score, 2), "%")
```

```python
title = "Decision Boundary"
xlabel = "Feature 1"
ylabel = "Feature 2"
plot_db_2d(nn, X_test, y_test, title, xlabel, ylabel, step=16)
```
